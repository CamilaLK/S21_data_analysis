```python
#Initiate the PCA and apply to your data
X = data_bce.iloc[:,0:47]
pca = sklearn.decomposition.PCA()
pca.fit(data_bce.iloc[:,0:47])
```

```python
#Apply the dimensionality reduction to our data
reduce_bce = pca.transform (X)
#Make a dataframe
reduce_bce = pd.DataFrame (reduce_bce)
reduce_bce
```

```python
#Plot the results of the reduction to look the dataframe
sns.scatterplot(data=reduce_bce, x=0, y=1)
plt.title ("PCA reduction")
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
sns.despine()
```

```python
#Finding the number os k
#Create a new inertia list and define a loop
Y = reduce_bce.iloc[:, 0:47]
inertia_list = []
for k in range (1,11):
    kmeans = sklearn.cluster.KMeans(k)
    kmeans = kmeans.fit(Y)
#Get the inertia of our initial kmean    
    inertia_list += [kmeans.inertia_]
#Check the inertia_list
inertia_list
```

```python
#Plot the cluster (k) with the inertia list
sns.scatterplot(x=np.arange(1,11), y=inertia_list)
sns.lineplot(x=np.arange(1,11), y=inertia_list)
sns.despine()
plt.title ("Elbow method")
plt.xlabel("k")
plt.ylabel("inertia")
```

# Based on the Elbow method, we can assume that there are probably 6 clusters for this dataframe

```python
#Quantity of clusters
k=6
```

```python
#Apply kmeans to our data
kmeans = sklearn.cluster.KMeans (k)
kmeans = kmeans.fit (Y)
```

```python
#Add the cluster column
reduce_bce ["cluster"] = kmeans.labels_
reduce_bce
```

```python
#Plot the columns 0 and 1 with hue cluster
sns.scatterplot(data=reduce_bce, x=0, y=1, hue="cluster")
sns.despine()
plt.title ("PCA reduction")
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
```

```python
#Plot the columns 0 and 1 in the different clusters
sns.relplot (data=reduce_bce, x=0, y=1, col="cluster")
sns.despine()
plt.title ("Different clusters in PCA")
plt.xlabel("PCA 0")
plt.ylabel("PCA 1")
```

# We can see that some cluster like 3, 4 and 5 have more condensed data than the cluster 0, 1 and 2. Cluster 3 and 4 have more values than the other ones.
